```python
score_set = []
while 1:
    print("환영합니다. 모의고사 성적관리 프로그램입니다.")
    print()
    print('   1.입력 2.출력 3.검색 4.수정 5.삭제')
    k = int(input('원하시는 기능의 번호를 입력해주세요.'))
    
    #1. 입력기능
    if k == 1:
        print()
        print('성적입력입니다.')
        name, kscore, escore, mscore = input('학생의 이름 / 국어성적 / 영어성적 / 수학성적 순서로 입력해주세요.').split()
        kscore = int(kscore)
        escore = int(escore)
        mscore = int(mscore)
        
        person_score = [name, kscore, escore, mscore, kscore+escore+mscore,(kscore+escore+mscore)/3 ]
        
        score_set.append(person_score)
        print('-----------------------------------')
        print(name,'의 국어성적: ', kscore)
        print(name,'의 영어성적: ', escore)
        print(name,'의 수학성적: ', mscore)
        print(name,'의 총점: ', kscore+escore+mscore)
        print(name,'의 과목평균: ', (kscore+escore+mscore)/3)
        print('-----------------------------------')
        
    #2. 출력기능
    if k == 2:
        print()
        print('출력기능입니다.')
        print('이름, 국어성적, 영어성적, 수학성적, 총점, 평균 순서대로 출력됩니다.')
        for i in range(len(score_set)):
            print(i+1,'.',' ', score_set[i][0],'/' ' ' '국어:',score_set[i][1], '/' ' ' '영어:',score_set[i][2], '/' ' ' '수학:',score_set[i][3], '/' ' ' '총점:',score_set[i][4], '/' ' ' '평균:',score_set[i][5])
            print()
            
    #3. 검색기능
    if k == 3:
        print() #엔터쳐라
        print('평균 점수 입력입니다.') #문구프린트
        
        avg = int(input('평균 점수를 입력해 주세요: ')) # 정수 입력
        print()
        print(avg,'에 해당하는 평균 점수 해당자 목록 입니다.') #설명출력
        print('-----------------------------------')
        
        for i in range(len(score_set)):
            #반복문 - score_set 원소의 개수만큼 반복
            if avg <= score_set[i][5]: #조건문 - 입력값과 각 학생의 평균점수 비교
                print('이름: ', score_set[i][0], '평균점수: ', score_set[i][5]) #조건 성립시 이름과 평균 점수 출력
                print()#i
            print()#n
        print() #1
        
    #4. 수정기능
    
    a = 0
    
    if k == 4:
        print()
        print('내용 수정 기능 입니다.')
        name = input('이름을 입력해 주세요: ') #이름 입력
        print(name, '에 해당하는 점수 정보 입니다.') #저장된 리스트에서 이름 값 검색
        print('-----------------------------------')
        print()
        for i in range(len(score_set)): #반복문 - score_set에서 원소의 개수만큼 반복
            if name == score_set[i][0]: #조건문 - 입력된 이름 값이 i는 0이다. 부터 시작해서, [0]번째 정보인 이름 값과 같으면
                print('이름: ', score_set[i][0], '/' ' ' '국어점수: ', score_set[i][1], '/' ' ' '영어점수: ', score_set[i][2], '/' ' ' '수학점수: ', score_set[i][3])
                print()

                kscore, escore, mscore = input('수정할 점수입력 - 국어성적 / 영어성적 / 수학성적 순서로 입력해주세요: ').split()

                kscore = int(kscore)
                escore = int(escore)
                mscore = int(mscore)

                score_set[i][1] = kscore
                score_set[i][2] = escore
                score_set[i][3] = mscore
                score_set[i][4] = kscore+escore+mscore
                score_set[i][5] = (kscore+escore+mscore)/3

                a += 1 # a+1 이며, if문 안에 있기 때문에 이름이 검색 되었을 때에만 실행 됨. 즉 a의 값은 1
 
        if a == 0: # 검색이 되지 않았기 때문에 a 는 0 이 된다.        
            print('해당하는 이름이 없습니다.')
            print()
        print('-----------------------------------')
               
        
    #5. 삭제기능

    a = 0
    
    if k == 5:
        print()
        print('내용 삭제 기능 입니다.')
        name = input('확인할 이름을 입력해 주세요: ')
        print(name, '에 해당하는 점수 정보 입니다.') #저장된 리스트에서 이름 값 검색
        print('-----------------------------------')
        print()
        
        for i in range(len(score_set)): #반복문 - score_set 원소의 개수만큼 반복
            if name == score_set[i][0]:
                print('이름: ', score_set[i][0], '/' ' ' '국어점수: ', score_set[i][1], '/' ' ' '영어점수: ', score_set[i][2], '/' ' ' '수학점수: ', score_set[i][3])
                print()
                
               

                name = input('삭제할 이름을 입력해주세요: ')
                
                score_set[i][0] = name
                score_set[i][0] = ' '
                

                a += 1
               
           
               
                
                
        if a == 0:
            print('해당하는 이름이 없습니다.')
            print()
        print('-----------------------------------')

                

         
```
